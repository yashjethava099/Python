## String

- Immutable sequence of characters

- **len(st)**
    
    Returns length of string
    
    Example: `len("Hello") → 5`
    
- **st.lower()** : Converts string to lowercase
    
- **st.casefold()** : Converts string to lowercase (more aggressive, used for comparisons)
    
- **st.upper()** : Converts string to uppercase
    
- **st.title()** : Converts first letter of each word to uppercase
    
- **st.capitalize()** : Converts first letter of string to uppercase
    
- **st.strip()** : Removes spaces from start and end
    
- **st.replace(old, new)** : Replaces one substring with another
    
- **st.find(value)** : Returns index of first occurrence, else -1
    
- **st.startswith(value)** : Checks if string starts with given value
    
- **st.endswith(value)** : Checks if string ends with given value
    
- **st.split()** : Splits string into list
    
- **"separator".join(list)** : Joins list elements into string
    
- **st.isalpha()** : True if all characters are alphabets
    
- **st.isdigit()** : True if all characters are digits
    
- **st.isalnum()** : True if characters are alphanumeric
    
- **st.zfill(width)** : Adds zeros at beginning
    
- **st.center(width)** : Aligns string to center

**Slicing**:

`st[start:end:step]`
        text="Python"

        print(text[0:3])# Pyt
        print(text[1:])# ython
        print(text[::-1])# nohtyP

## Tuple

- Ordered, immutable

**Methods**:

    1. count() : Returns the number of times a value occurs in the tuple
        tup= (1,2,2,3)
        print(tup.count(2))# Output: 2
    2. index() : Returns the index (position) of the first occurrence of a value
        tup= (10,20,30)
        print(tup.index(20))# Output: 1

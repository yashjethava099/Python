<h1>3. Core Python Concepts</h1>

1. Data types: Python has built-in data types like integers (whole numbers), floats (decimal numbers), strings (text), lists (ordered & changeable), tuples (ordered & unchangeable), dictionaries (key-value pairs), and sets (unordered unique items).

2. Variables & memory: Variables are names that store data by referencing objects in memory, and Python automatically manages memory allocation.

3. Operators: Python includes arithmetic operators (+, -, *, /), comparison operators (==, !=, >, <), logical operators (and, or, not), and bitwise operators (&, |, ^) to perform different operations.

```python
# Demonstrating variables and data types in Python

# Integer
age = int(input("Enter your age: "))

# Float
height = 5.9

# String
name = input("Enter your name: ")

# List
marks = [85, 90, 78]

# Tuple
coordinates = (10, 20)

# Dictionary
student = {"name": "Yash", "age": 21}

# Set
unique_numbers = {1, 2, 3, 3}

# Printing values and their data types
print("Name:", name, "| Type:", type(name))
print("Age:", age, "| Type:", type(age))
print("Height:", height, "| Type:", type(height))
print("Marks:", marks, "| Type:", type(marks))
print("Coordinates:", coordinates, "| Type:", type(coordinates))
print("Student:", student, "| Type:", type(student))
print("Unique Numbers:", unique_numbers, "| Type:", type(unique_numbers))
```

<h1>4. Conditional Statements</h1>

1. Conditional statements in Python are used to perform different actions based on different conditions. The main types are:

        if: Executes code when a condition is true
        elif: Checks another condition if the previous one is false
        else: Executes code when all conditions are false
2. Nested if-else means using an if or else statement inside another if or else block. It is used when multiple conditions need to be checked step by step.

```python
#Practical Example 5: Write a Python program to find greater and less than a number using if_else.
num1 = 10
num2 = 20
if num1 > num2:
    print("num1 is greater than num2")
else:
    print("num1 is not greater than num2")
```

```python
# Practical Example 6: Write a Python program to check if a number is prime using if_else.
num = 10
if num > 1:
    for i in range(2, int(num/2) + 1):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
```

```python
# Practical Example 7: Write a Python program to calculate grades based on percentage using if-else ladder.
percentage = 85
if percentage >= 90:
    print("Grade: A")
elif percentage >= 80:
    print("Grade: B")
elif percentage >= 70:
    print("Grade: C")
elif percentage >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
# Practical Example 8: Write a Python program to check if a person is eligible to donate blood using a nested if.
age = 25
weight = 50

if age >= 18:
    if weight > 45:
        print("Person is eligible to donate blood")
    else:
        print("Person is not eligible to donate blood due to low weight")
else:
    print("Person is not eligible to donate blood due to age")
```

<h1>5. Looping (For, While)</h1>

1. Loops in Python are used to repeat a block of code multiple times.

        for loop is used to iterate over a sequence like list, tuple, or string
        while loop runs as long as a condition is true

2. Loops execute code repeatedly based on a condition or sequence.

        In a for loop, the loop runs for each item in a collection
        In a while loop, the loop continues until the condition becomes false

3. Loops can be used with collections like lists, tuples, and strings to access each element one by one.
This makes it easy to process or display data from collections.

```python
# Practical Example 1: Write a Python program to print each fruit in a list using a simple for loop. List1 = ['apple', 'banana', 'mango']
fruits = ['apple', 'banana', 'mango']
for i in fruits:
    print(i)
```

```python
# Practical Example 2: Write a Python program to find the length of each string in List1.
list1 = ['apple', 'banana', 'mango']
for i in list1:
    print(f"Length of '{i}': {len(i)}")
```

```python
# Practical Example 3: Write a Python program to find a specific string in the list using a simple for loop and if condition.
list1 = ['apple', 'banana', 'mango']
search_string = 'banana'
for i in list1:
    if i == search_string:
        print(f"'{search_string}' found in the list.")
        break
else:
    print(f"'{search_string}' not found in the list.")
```

```python
# Practical Example 4: Print this pattern using nested for loop:
# *
# **
# ***
# ****
# *****
lines = 5
for i in range(1, lines+1):
    print('*'*i)
```

```python
def square(x):
    square = x * x
    print(f"square of {x} is {square}")
square(5)
```

```python
def add(x, y):
    sum = x + y
    print(f"sum of {x} and {y} is {sum}")
add(5, 10)
```

```python
def mul(x, y):
    r = x * y
    return r
result = mul(5, 10)
print(f"result is {result}")
```

```python
def person(name, email, phone):
    print(name, email, phone)
person("Yash", "yash@gmail.com", 1234567890)
```

```python
def person(name, email, phone=9898989898):
    print(name, email, phone)
person("Yash", "yash@gmail.com")
```

```python
def add(*a):
    print(a)
add(1, 2, 3, 4, 5)
```

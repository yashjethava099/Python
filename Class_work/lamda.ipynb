```python
l = ["python", "java", "c++", "ruby"]
k = map(lambda x: len(x), l)
print(list(k))
```

```python
l = ["python", "java", "c++", "ruby"]
k = filter(lambda x: "a" in x, l)
print(list(k))
```

```python
#startwith p
l = ["python", "java", "c++", "ruby"]
k = filter(lambda x: x.startswith("p"), l)
print(list(k))
```

```python
#print only perfect square root in filter
l = [2,5,9,8,5,49,45,81,144,22,33]
k = filter(lambda x: int(x**0.5)**2 == x, l)
print(list(k))
```

```python
#find max no

from functools import reduce

l = [10, 20, 30, 40, 50]
def max_no(x, y):
    if x > y:
        return x
    else:
        return y
k = reduce(max_no, l)
print(k)
```

```python
from functools import reduce
l = [10, 20, 30, 40, 50]
k = reduce(lambda x, y: x if x>y else y, l)
print(k)
```

```python
#generator
def gen(n):
    for i in range(1,n):
        yield i*i
k = gen(10)
print(next(k))
print(next(k))
print(next(k))
print(next(k))
print(next(k))
print(next(k))
print(next(k))
```

```python
#generator of factorial
def gen(n):
    f = 1
    for i in range(1, n+1):
        f *= i
        yield f
k = gen(10)
print(next(k))
print(next(k))
print(next(k))
print(next(k))
print(next(k))
print(next(k))
```

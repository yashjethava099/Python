```python
lines = 5
for i in range(lines):
    for j in range(i+1):
        print(j+1,end="")
    print()
```

```python
lines = 5
star = 1
c=lines
for i in range(lines):
    for j in range(star):
        print(c+j,end="")
    star+=1
    c-=1
    print()
```

```python
lines = 5
star = 1
for i in range(lines):
    for j in range(star):
        print((i+j)%2,end="")
    star+=1
    print()
```
